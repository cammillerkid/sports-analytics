```python
agg = df.loc[(df['mpg']>=10) & (df['gp']>10)]

agg.groupby(['all_star']).agg({
                                    'pos':'count',                         
                                    'age':['mean','median','min','max'],
                                    'gp':['mean','min'],
                                    'usage_rate':'mean',
                                    'points_per_game':'mean',
                                    'versatility_index':'mean',
                                    'off_rating':'mean',
                                    'def_rating':'mean'
                                    })
```

```python

cols = ['mpg',
        'gp',
        'usage_rate',
        'assists_per_game',
        'points_per_game',
        'off_rating',
        'def_rating',
        'versatility_index',
        'steals_per_game',
        'turnover_per_game',
        'blocks_per_game'
       ]
```

```python
values = []
suffix = '_90th'

for x in cols:
    vals = x+suffix
    values.append(vals)
```

```python
cols_dict = dict(zip(cols, values))

cols_dict
```

```python
for k, v in cols_dict.items():
    df[v] = np.percentile(df[k],90) # gives the 95th percentile
```

```python
df['mpg_better'] = np.where(df['mpg']>=df['mpg_90th'],1,0)
df['gp_better'] = np.where(df['gp']>=df['gp_90th'],1,0)
df['usage_rate_better'] = np.where(df['usage_rate']>=df['usage_rate_90th'],1,0)
df['assists_per_game_better'] = np.where(df['assists_per_game']>=df['assists_per_game_90th'],1,0)
df['points_per_game_better'] = np.where(df['points_per_game']>=df['points_per_game_90th'],1,0)
df['off_rating_better'] = np.where(df['off_rating']>=df['off_rating_90th'],1,0)
df['def_rating_better'] = np.where(df['def_rating']>=df['def_rating_90th'],1,0)
df['versatility_index_better'] = np.where(df['versatility_index']>=df['versatility_index_90th'],1,0)
df['steals_per_game_better'] = np.where(df['steals_per_game']>=df['steals_per_game_90th'],1,0)
df['turnover_per_game_better'] = np.where(df['turnover_per_game']>=df['turnover_per_game_90th'],1,0)
df['blocks_per_game_better'] = np.where(df['blocks_per_game']>=df['blocks_per_game_90th'],1,0)
```
